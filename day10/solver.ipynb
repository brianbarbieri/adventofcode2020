```python
with open("data.txt", "r") as r:
    data = [int(l.replace("\n", "")) for l in r.readlines()]
```

```python
# part 1
device_jolt = max(data) + 3
start_jolt = 0

ones = 0
threes = 0

while True:
    if start_jolt + 1 in data:
        start_jolt += 1
        ones += 1
    elif start_jolt + 3 in data:
        start_jolt += 3
        threes += 1
    else:
        print(f"Output is ones: {ones}, threes: {threes+1} and answer is {ones*(threes+1)}")
        break
```

```python
# part 2  # pretty slow, should be a suboptimal solution.
import numpy as np 

data = np.array(sorted(data))
max_value = data[-1]

class Branch:

    def __init__(self, value):
        self.value = value

    def find_subbranches(self, data):
        self.subbranches = data[(data > self.value) & (data <= self.value+3)]
i = 0
completed_branches = 0
to_traverse = [Branch(0)]
while len(to_traverse) != 0:
    branch = to_traverse.pop()
    branch.find_subbranches(data)
    for b in branch.subbranches:
        if b == max_value:
            completed_branches += 1
        else:
            to_traverse.append(Branch(b))
    
    i += 1
    if i % 100 == 0:
        print(len(to_traverse), completed_branches, branch.value, max_value)

print(f"There are {completed_branches} arrangements found")
```
